#### Задание 24. Визуализировать данные вашего скаченного DataSet в стиле Столбчатые диаграммы (подписать оси, заголовок, столбцы зеленого цвета)

#### Задание 25. Визуализировать данные вашего скаченного DataSet  в стиле Точечный график (подписать оси, заголовок, столбцы красного цвета)

```python
X_array = cybersecurity_attacks_scores['Timestamp'].values
Y_array = cybersecurity_attacks_scores['Anomaly Scores'].values
plt.scatter(X_array, Y_array, color = 'r')
plt.title("Точечный график")
plt.xlabel('Время')
plt.ylabel('Показатель аномалии атаки')
plt.grid()
```

#### Задание 26. Визуализировать данные вашего скаченного DataSet в стиле Линейный график (подписать оси, заголовок, цвет, стиль линии изменить на ваше усмотрение)

```python
X_array = cybersecurity_attacks_scores['Timestamp'].values
Y_array = cybersecurity_attacks_scores['Anomaly Scores'].values
line = plt.plot(X_array, Y_array, color = 'c')
plt.setp(line, linestyle = '--')
plt.title("Линейный график")
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
```

#### Задание 27. Визуализировать данные вашего скаченного DataSet в стиле Цветовое распределение

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()


for column in cybersecurity_attacks_scores.columns:
    cybersecurity_attacks_scores[column] = encoder.fit_transform(cybersecurity_attacks_scores[column])


plt.pcolor(cybersecurity_attacks_scores)
plt.colorbar()
```
